```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
df=pd.read_excel("Copy of Dormant Account Survey form (Responses).xlsx")
df2=df.loc[df['Gender'] == "Female", 'Education qualification']
lst=list(df2)
count=0;
count2=0;
for i in lst:
    if(i=="Graduate"):
        count=count+1
print(count)
df3=df.loc[df['Gender'] == "Male", 'Education qualification']
lst2=list(df3)
for x in lst2:
    if(x=="Graduate"):
        count2+=1
print(count2)
female_graduates= (count/89)*100
male_graduates= (count2/89)*100
lst3 = [female_graduates, male_graduates]
label = ['female graduates' , 'male graduates']
exp= [0.1, 0]
plt.pie(lst3 ,labels = label ,explode = exp, shadow = True , radius = 1, autopct = "%1.2f%%")
plt.title("female graduates vs male graduates" , color = "black")
plt.show()


```

```python
## data cleansing using map() function
## cleansing data of house distance column
df['House Distance from Bank(in KM)'] = df["House Distance from Bank(in KM)"].map(str)
```

```python
df['House Distance from Bank(in KM)'].apply(type)
```

```python
df['House Distance from Bank(in KM)'] = df['House Distance from Bank(in KM)'].str.replace('kms', '')
df['House Distance from Bank(in KM)'] = df['House Distance from Bank(in KM)'].str.replace('km', '')
df['House Distance from Bank(in KM)'] = df['House Distance from Bank(in KM)'].str.replace('m', '')
df['House Distance from Bank(in KM)'] = df['House Distance from Bank(in KM)'].str.replace('KM', '')
df['House Distance from Bank(in KM)']
```

```python
## removing extra spaces from account num column and and house distance column
df['House Distance from Bank(in KM)'] = df['House Distance from Bank(in KM)'].str.replace(' ', '')
df['House Distance from Bank(in KM)']
```

```python
df['Account NO(not manadatory to disclose)'] = df['Account NO(not manadatory to disclose)'].str.replace(' ', '')
```
